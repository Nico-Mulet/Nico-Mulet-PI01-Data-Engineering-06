Creamos una nueva columna llamada "id" sumado a la columna show_id para asi formar una id para cada plataforma por ejemplo "as123"

```python
df_amazon.insert(loc=0, column="id", value="a" + df_amazon["show_id"])
df_disney_plus.insert(loc=0, column="id", value="d" + df_disney_plus["show_id"])
df_hulu.insert(loc=0, column="id", value="h" + df_hulu["show_id"])
df_netflix.insert(loc=0, column="id", value="n" + df_netflix["show_id"])
```

Juntamos todos los dataframe en uno solo para despues poder trabajar con ellos 

```python
df_plataformas = df_amazon.append(df_disney_plus).append(df_hulu).append(df_netflix)
```

Remplazamos los valones nulos de la columna rating por la letra g que esto significa "General for all audiences"

```python
df_plataformas["rating"].fillna("g", inplace=True)
```

Verificamos los cambios

```python
df_plataformas.head()
```

Cambio de formato de fecha a la columna "date_added" en AAAA/MM/DD

```python
df_plataformas["date_added"] = pd.to_datetime(df_plataformas["date_added"])
```

Verificamos los cambios

```python
df_plataformas.head(5)
```

Cambiamos todas las letras mayusculas en minusculas y verificamos los cambios

```python
df_plataformas["type"] = df_plataformas["type"].str.lower()
df_plataformas["title"] = df_plataformas["title"].str.lower()
df_plataformas["director"] = df_plataformas["director"].str.lower()
df_plataformas["cast"] = df_plataformas["cast"].str.lower()
df_plataformas["country"] = df_plataformas["country"].str.lower()
df_plataformas["listed_in"] = df_plataformas["listed_in"].str.lower()
df_plataformas["description"] = df_plataformas["description"].str.lower()
df_plataformas["rating"] = df_plataformas["rating"].str.lower()
df_plataformas["duration"] = df_plataformas["duration"].str.lower()
df_plataformas
```

A la columna "duration" la separamos en dos columnas. Una en "duration_type" donde alli entraria los "min" y los "season" y la otra columna "duration_int" donde estaria las duraciones y verificamos los cambios

```python
separador = df_plataformas["duration"].str.split(" ", n=1, expand=True)
df_plataformas["duration_int"] = separador[0]
df_plataformas["duration_type"] = separador[1]
df_plataformas["duration_type"] = df_plataformas["duration_type"].str.replace("seasons", "season")
df_plataformas["duration_int"] = df_plataformas["duration_int"].astype("Int64")
df_plataformas.head(10)
```

Guardamos el dataframe 

```python
df_plataformas.to_csv("plataformas_database.csv", index=False)
```
